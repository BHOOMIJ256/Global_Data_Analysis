### **MISSING VALUE IMPUTATION (BY MEAN)**

```python
data_cleaned = data.dropna(thresh=int(0.7 * len(data)), axis=1)
```

```python
numeric_cols = data_cleaned.select_dtypes(include='number').columns
data_cleaned[numeric_cols] = data_cleaned[numeric_cols].fillna(data_cleaned[numeric_cols].mean())
```

### **GROSS DOMESTIC PRODUCT ANALYSIS**

```python
#GDP ANALYSIS COUNTRY WISE


import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np

#min value of gdp
min = np.min(data_cleaned['gdp'])
print("Minimum GDP value is :",min)

#max value of gdp
max = np.max(data_cleaned['gdp'])
print("Maximum Value of GDP is ",max)

#mean value of gdp
mean = np.mean(data_cleaned['gdp'])
print("Mean Value of GDP is :",mean)

#median value of gdp
median = np.median(data_cleaned['gdp'])
print("Median value of GDP is :",median)
```


**1)The minimum value of GDP is 46 (in current US Dollar)**

**2) The maximum value of GDP in the distribution is 20580223.0 (US Dollars)**

**3) The average value or the mean value of the distribution is 445130 (US Dollars)**


```python
with pd.option_context('mode.use_inf_as_na', True):
    # Drop rows with NaN (which includes inf values after conversion)
    data = data.dropna(subset=['unemployment'])

#Distribution of gdp values
plt.figure(figsize=(15,5))
sns.histplot(data_cleaned['gdp'],kde=True)
plt.title("DISTRIBUTION OF GDP")
plt.xlabel("GDP")
plt.show()


#Countries with Highest GDP
top_gdp = data_cleaned.sort_values(by="gdp",ascending = False).head(10)
fig = plt.figure(figsize =(15,5))
sns.barplot(x ="name",y ="gdp",data = top_gdp,palette = "Blues")
plt.xlabel("Countries")
plt.ylabel("GDP(US Dollars)")
plt.title("COUNTRIES WITH HIGHEST GDP")
plt.show()

#countries with lowest GDP
low_gdp = data_cleaned.sort_values(by="gdp",ascending= True).head(10)
fig = plt.figure(figsize =(20,5))
sns.barplot(x ="name",y ="gdp",data = low_gdp , palette= "Oranges")
plt.xlabel("Countries")
plt.ylabel("GDP(US Dollars)")
plt.title("COUNTRIES WITH LOW GDP")
plt.show()
```

**Inference From the First Diagram**

**1) The GDP distribution is heavily skewed to the right. This indicates that the majority of the countries have relatively low GDP values, while a few countries have significantly higher GDP values, creating a long tail to the right.

2) The right tail with only a few countries indicates the presence of outliers with extremely high GDP. These outliers likely represent economically dominant countries or large economies.

**Inference From the Second Diagram**

1) The United States has the highest GDP, followed by China, indicating their positions as the largest economies in the world.

2) There is a significant gap between the GDP of the United States and the other top countries, showing the strong economic dominance of the U.S.

3) The values of GDP decrease gradually as we move from the United States down to Brazil, the last country shown in the highest-GDP chart.

**Inference From Third Diagram**

1) On the lower end of the GDP spectrum, countries like Tuvalu, Nauru, and Kiribati have the smallest GDP figures.

2) These countries with low GDP are mostly smaller nations or island states, often with limited natural resources or small populations, which can restrict their economic capacity.

### **EMPLOYMENT AND UNEMPLOYMENT**

```python
#UNEMPLOYMENT COUNTRY-WISE


import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

fig = plt.figure(figsize =(10,5))
sns.histplot(data_cleaned['unemployment'],kde =True)
plt.xlabel("Unemployment in Percentage")
plt.legend()
plt.show()

#TOP 10 COUNTRIES WITH HIGHEST UNEMPLOYMENT
top_unemployed_countries = data_cleaned.sort_values(by="unemployment",ascending= False).head(10)
fig = plt.figure(figsize = (10, 5))
sns.barplot(x='unemployment', y='name', data=top_unemployed_countries, palette="Reds")
plt.title("Top 10 Countries with highest employment")
plt.xlabel("Unemployment")
plt.ylabel("Country")

plt.show()

#TOP 10 COUNTRIES WITH LOWEST UNEMPLOYMNET
least_unemployed = data_cleaned.sort_values(by="unemployment").head(10)
fig = plt.figure(figsize =(10,5))
sns.barplot(x="unemployment",y="name",data=least_unemployed,palette ="Greens")
plt.title("Top 10 Countries with lowest unemployment")
plt.xlabel("Unemployment")
plt.ylabel("Country")
plt.show()


#percentage of employment in industry
fig = plt.figure(figsize =(10,5))
sns.histplot(data = data_cleaned , x ="employment_industry")
plt.xlabel("Employment in Industrial Sector")
plt.ylabel("Count")
plt.title("Distribution of Employment in Industrial Sector")
plt.show()

#percentage of employment in agriculture sector
fig = plt.figure(figsize =(10,5))
sns.distplot(a=data_cleaned['employment_agriculture'],hist = False,kde = True , color ="Black")
plt.xlabel("Employment in Agriculture Sector")
plt.ylabel("Count")
plt.title("Distribution of Employment in Agricultural Sector ")
plt.show()

#percentage of employment in service sector
fig = plt.figure(figsize =(10,5))
sns.stripplot(x = data_cleaned["employment_services"],color ="Orange")
plt.xlabel("Employment in Service Sector")
plt.ylabel("Count")
plt.title("Distribution of Employment in Service Sector")
plt.show()

#3D SCATTER PLOT FOR employment in all three sectors
fig = px.scatter_3d(data_cleaned, x ="employment_services",y="employment_industry",z ="employment_agriculture")
fig.show()

```

**Inference From the First Diagram**

The distribution appears to be right-skewed, with most unemployment rate
clustered around the lower range (0-5%), and fewer countries experiencing higher unemployment rates (>10%).

The highest count of countries seems to have an unemployment rate of around 3-6%.

A smaller number of countries have unemployment rates extending up to 30%, indicating some outliers or extreme values.

The preciding 2 diagrams show the top 10 countries with highest unemployment and top 10 countries with lowest unemployment

**Inference For Employment in Industry**

It can be observed that the distribution for employment in industrial sector concentrates maximum around 10-30%.
With Few Outlier countries with more than 40% employment in industrial sector

**Inference For Employment in Agricultural Sector**

The distribution is observed to be right-skewed , indicating maximum employment rates clustered around 0-40% , with fewer countries experiencing higher employment rates greater than >40%.


**Inference For Employment in Service Sector**

It is observed from the distribution that maximum data points ranges from 40%-80%. With fewer countries having less than 40% employed in Service Sector,and  outlier country having more than 80% employed in Service Sector
